```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding="latin-1")
pd.DataFrame(df)
```

```python
year_17 = df.groupby("Area")["Y2017"].sum().reset_index()
print(year_17)
max_year = year_17[year_17["Y2017"] == year_17["Y2017"].max()]
print(max_year)
min_year = year_17[year_17["Y2017"] == year_17["Y2017"].min()]
print(min_year)
```

```python
df["Y2017"].describe()
```

```python
missing_sum = df["Y2014"].isnull().sum()
total_count = df["Y2014"].count()
print (f"The total number of missing values in Y2014 is {missing_sum} and the percentage of missing values is {missing_sum/total_count*100}%")
```

```python
condition = (df["Area"] == "Madagascar") & (df["Element"] == "Protein supply quantity (g/capita/day)")
supply = df.loc[condition, "Y2015"].sum()
#print((condition==True).sum())
print(f"The total Protein supply quantity in Madagascar in 2015 is: {supply}")
```

```python
sorted_year_17 = year_17.sort_values(by="Y2017", ascending=True)
seventh_lowest = sorted_year_17.iloc[6,0]
print(f"{seventh_lowest} had the seventh lowest sum in 2017")
```

```python
element_group = df.groupby("Element")["Y2017"].sum().reset_index()
processing = element_group.iloc[10,1]
print(f"The total number of the sum of Processing in 2017 is: {processing}")
```

```python
all_group = df.groupby("Element")
y_14= all_group["Y2014"].sum().reset_index()
y_15= all_group["Y2015"].sum().reset_index()
y_16= all_group["Y2016"].sum().reset_index()
y_17= all_group["Y2017"].sum().reset_index()
y_18= all_group["Y2018"].sum().reset_index()
stock = [y_14.iloc[15,], y_15.iloc[15,], y_16.iloc[15,], y_17.iloc[15,], y_18.iloc[15,]]
print(stock)
print(f"Year 2014 had the highest total stock variation with a value of {y_14.iat[15,1]}")
```
